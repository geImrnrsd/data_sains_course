```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
gb = df.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].agg("mean")
gb
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
import numpy as np

labels = gb.columns
x = np.arange(len(labels))

with plt.style.context('ggplot'):
  fig, ax = plt.subplots(figsize=(16,7))
  width = 0.07
  for idx, label in enumerate(gb.index):
    y = gb.loc[label].values
    offset = (idx - len(gb.index)/2)*width 
    xt = x + offset
    ax.bar(xt, y, width, label=label)

ax.set_title("Mean Sales Video Games By Genre")
ax.set_ylabel("Mean Sales")
ax.set_xlabel("Region Sales")

plt.xticks(x, labels)
plt.legend(facecolor="#E5E5E5", edgecolor="silver", bbox_to_anchor=(0.45, 0.56, 0.6, 0.6))
plt.show()
```

Jawab:

1. Pada grafik kita bisa melihat distribusi penjualan games tiap region berdasarkan genrenya. Sehingga kita bisa melihat minat orang terhadap genre tiap wilayah berdasarkan jumlah pembelian.
  - Misalnya Pada NA_Sales kita bisa melihat bahwa "Platform" memiliki peminat yang paling banyak sedangkan "Adventure" memiliki peminat yang paling sedikit. 
  - Contoh lain, pada "JP_Sales", orang-orang lebih minat dengan permainan "Role Playing" 

2. Selain itu, kita bisa membandingkan penjualan "genre" tiap wilayah. 
  - Misalnya, peringkat penjualan games genre "Action" berturut2: NA_Sales, EU_sales, Other_Sales, JP_Sales
3. Kita juga bisa melihat kecenderungan penjualan category games pada tiap wilayah. Hal ini bisa dijadikan acuan untuk menentukan game dengan genre apa. 
  - Misal, penjualan games bergenre "platform" cukup menjanjikan, bisa dilihat dari grafik, bahwa penjualan kategori tersebut pada tiap region cukup baik. 
  - Kemudian kita juga bisa melihat penjualan game genre "adventure" memiliki peminat yang sedikit tiap wilayah. 

4. Meskipun grafik ini fokus terhadap penjualan game tiap wilayah berdasarkan "genre", kita juga bisa melihat akumulatif penjualan tiap wilayah.
  - Dari grafik nampak bahwa "NA_Sales" secara akumulatif memiliki penjualan yang terbaik dibandingkan yang lain pada setiap genrenya. Sedangkan "Other_Sales" adalah penjualan terburuk.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
labels = gb.columns
x = np.arange(len(labels))

with plt.style.context('ggplot'):
  fig, ax = plt.subplots(figsize=(14,7))
  bottom = [0 for i in range(len(labels))]
  for idx, label in enumerate(gb.index):
    y = gb.loc[label].values
    ax.bar(x, y, bottom=bottom, label=label)
    bottom += y

ax.set_title("Mean Sales Video Games By Genre")
ax.set_ylabel("Mean Sales")
ax.set_xlabel("Region Sales")

plt.xticks(x, labels)
plt.legend(facecolor="#E5E5E5", edgecolor="silver", bbox_to_anchor=(0.45, 0.56, 0.6, 0.6))
plt.show()
```

Jawab:
1. Pada grafik di atas, kita bisa melihat total dari rata-rata penjualan games pada setiap region dengan ditambahkan rincian perkategori.

  - Urutan penjualan games dari terbaik ke terburuk adalah : NA_Sales, EU_Sales, JP_Sales, dan Other_Sales. 

2. Meski bentuk grafik ini fokus terhadap total penjualan, namun kita bisa melihat Rasio penjualan perkategori tiap region. Hal ini bisa dilihat dari "tinggi" dari bar dari kategori. Semakin tinggi warna pada kategori, semakin tinggi pula rasionya. 
  - Misalnya pada "NA_Sales" dapat dilihat bahwa penjualan games bergenre "platform" memiliki penjualan tertinggi dan "adventure" dengan penjualan terendah
