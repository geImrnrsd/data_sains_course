```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
ys = [
    {"data" : tan, "title":"Grafik tanx", "label": "tanX", "color":"b"},
    {"data": sin, "title":"Grafik sinx", "label": "sinX", "color":"r"}, 
    {"data": cos, "title":"Grafik cosx", "label": "cosX", "color":"g"}, 
]

fig, ax = plt.subplots(1, 3, figsize=(18,8), sharey=True)

lines = []
labels = []
for idx, obj in enumerate(ys):
  ax[idx].set_title(obj.get("title"))
  line = ax[idx].plot(
      x, 
      obj.get("data") , 
      color=obj.get("color")
  )[0]
  lines.append(line)
  labels.append(obj.get("label"))

fig.legend(
    handles=lines,
    labels=labels,
    loc='upper center'
)

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(14, 8))

width = 0.3
height = 0.3
ys = [
      {"data": y2, "title":"Grafik CosX^2", "color": "r", "position": [-0.1, -0.1, width, height]},
      {"data": y3, "title":"Grafik CosX^3", "color": "g", "position": [0.9, -0.1, width, height]},
      {"data": y4, "title":"Grafik CosX^4", "color": "k", "position": [0.9, 0.9, width, height]},
      {"data": y5, "title":"Grafik CosX^5", "color": "y", "position": [-0.1, 0.9, width, height]}, 
]

# Main Graph
ax.plot(x, y)

ins = [None for i in range(4)]

for idx, obj in enumerate(ys):
  ins[idx] = ax.inset_axes(obj.get("position"))
  ins[idx].set_title(obj.get("title"))
  ins[idx].plot(x, obj.get("data"), color=obj.get("color"))
  ins[idx].plot(x, y, color="b")

plt.title("Grafik CosX")
plt.show()
```
