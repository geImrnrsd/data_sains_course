<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
Perbedaan Series dengan List
- Series hanya dapat menyimpan data yang homogen (satu tipe data), sedangkan pada tipe data List kita bisa menyimpan data yang heterogen (berbeda tipe data).
- List adalah built in data type pada python, sedangkan Series merupakan kelas dari pandas
- index pada list berupa nilai integer yang terurut(mulai dari 0 sampai n-1), Sedangkan pada series index bisa berupa nilai string yang unik    
- terdapat perbedaan hasil/definisi operasi + , -, *, / 

Perbedaan Series dengan Dictionary
- pada series kita bisa mengimplementasikan metode slicing sedangkan pada dictionary tidak bisa.
- Dictionary adalah built in data type pada python, sedangkan Series merupakan kelas dari pandas
- terdapat perbedaan hasil/definisi operasi + , -, *, / 

Perbedaan dataframe dan series
- series merupakan data satu dimensi sedangkan dataframe merupakan data 2-dimensi
- jika diibaratkan pada tabel, series merupakan kolom pada tabel tsb sedangkan dataframe adalah keseluruhan tabel

```python
# code here
data = ["jhon", "doe", "jane", "tim"]
sr = pd.Series(data)
sr
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
description = df2.describe()
description
```

```python
print("Berdasarkan tabel di atas diperoleh data sebagai berikut:")
data = [
  {"name":"price", "label":"rata-rata", "key":"mean"},
  {"name":"depth", "label":"standar deviasi", "key":"std"},
  {"name":"carat", "label":"maksimum", "key":"max"}
]

for obj in data:
  name = obj.get("name")
  label = obj.get("label")
  key = obj.get("key")
  value = round(description[name][key],6)
  print(f"- Nilai {label} dari kolom \"{name}\" adalah {value} ")
```

Jawab: 
Berdasarkan tabel di atas, diperoleh data sebagai berikut:
- Nilai rata-rata dari kolom "price" : 3932.799722
- Nilai standar deviasi dari kolom "depth" : 1.432621
- Nilai maksimum dari kolom "carat" : 5.010000
