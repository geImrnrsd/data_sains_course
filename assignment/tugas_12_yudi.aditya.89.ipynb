Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

ys = [
    {"data": cos, "color":"r", "label":"Cosx"}, 
    {"data": cos3, "color":"g", "linestyle":"dashed", "label":"3 Cosx"}, 
    {"data": sin, "color":"b", "linestyle":"dashdot", "label":"Sin(x-90)" }, 
    {"data": sin2, "color":"y", "linestyle":"none", "marker":"x", "label":"2 Sinx"}, 
]

fig, ax = plt.subplots(figsize=(9.5,7))

def parseValue (obj, key): 
  return obj.get(key) if obj.get(key)!=None else None

for y in ys: ax.plot(
      x, 
      y.get("data"), 
      color=y.get("color"), 
      linestyle=parseValue(y,"linestyle"), 
      marker=parseValue(y,"marker"), 
      label=parseValue(y,"label")
  )

ax.legend(loc="center")

plt.title("Grafik Cos Sin")
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")

plt.show()
```

```python
class MultiPlot:
  def __init__(self, title, x):
    self.fig, self.ax = plt.subplots(figsize=(9.5,7))
    self.x = x
    self.title = title
    self.xLabel = "X-axis"
    self.yLabel = "Y-axis"
  
  def __parseValue(self,obj):
    return obj.get(key) if obj.get(key)!=None else None

  def setLabel(self,xLabel,yLabel):
    self.xLabel = xLabel
    self.yLabel = yLabel

  def appendLine(self, obj):
    self.ax.plot(
        self.x, 
        obj.get("data"),
        color=obj.get("color"), 
        linestyle=parseValue(obj,"linestyle"), 
        marker=parseValue(obj,"marker"), 
        label=parseValue(obj,"label")
    )
  
  def show(self):
    self.ax.legend(loc="center")
    plt.xlabel(self.xLabel)
    plt.ylabel(self.yLabel)
    plt.title(self.title)
    plt.plot() 

mPlot = MultiPlot("Grafik Cos Sin",x)
mPlot.setLabel("X-Axis Label","Y-Axis Label")
for y in ys: mPlot.appendLine(y)
mPlot.show()
```
