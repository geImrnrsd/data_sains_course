<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# Buatlah Figure dan Axis kosong tanpa Data
fig, ax = plt.subplots()
print("Menampilkan figure kosong")
plt.show()
```

```python
# Plot data secara Eksplisit 
print("Plot Data Secara Eksplisit")
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()

# Plot data secara Implisit
print("Plot Data Secara Implisit")
plt.plot(x,y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
# Generate data
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)

ys = [
      {"data": y, "color":"#1f77b4"}, 
      {"data": y2, "color": "#2ca02c"}, 
      {"data": y3, "color": "#d62728"}
     ]

# plt.style.use('seaborn')
with plt.style.context('seaborn'):
  fig, ax = plt.subplots()
  for y in ys: ax.plot(x,y.get("data"), color=y.get("color"))

# Mencegah menampilkan plot pada console
print("Menampilkan hasil plot dengan style seaborn")
plt.show()

# menyimpan hasil plot dalam bentuk gambar
fileName = "soal-3.png"
fig.savefig(fileName)

# menampilkan gambar yang telah disimpan
from IPython.display import Image
print("\n\nMenampilkan gambar", fileName)
Image(fileName)
```
