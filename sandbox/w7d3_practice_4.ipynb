```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://nasional.kompas.com/read/2020/10/15/18022671/polri-ungkap-percakapan-para-tersangka-di-grup-whatsapp-kami-medan?page=all#page2"

response = requests.get(url)
doc = BeautifulSoup(response.text, 'html.parser')
print(doc)
```

```python
box = doc.find("div", class_="read__content")
print(box)
```

```python
linkTags = doc.findAll('strong')

for tag in linkTags:
    # remove a tag in linkTag
    tag.decompose()    
```

```python
paragraphTags = doc.findAll('p')
result = [paragraph.get_text() for paragraph in paragraphTags]

print(result)
```

```python
# Cleaning data

import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def cleanTexts(texts):
    result = []
    for text in texts:
        text = text.lower()
        text = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())
        result.append(text)
    return result

result = cleanTexts(result)
paragraph = " ".join(result)
stop_words = set(stopwords.words('indonesian'))
word_tokens = word_tokenize(paragraph)
paragraph2 = " ".join([w for w in word_tokens if not w in stop_words])
print(paragraph2)
```
